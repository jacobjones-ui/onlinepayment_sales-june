```python
#Drop Rows with Missing Key Data (if needed)
df.dropna(subset=["Product Name", "Total Amount", "Qty"], inplace=True)
df
```

```python
df.columns = df.columns.str.strip()
df
```

```python
# Top 10 Best Selling Products by Total Amount
top_products_amount = df.groupby("Product Name")["Total Amount"].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Total Sales Amount:")
print(top_products_amount)
```

```python
 #Top 10 Products by Quantity Sold
top_products_qty = df.groupby("Product Name")["Qty"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products by Quantity Sold:")
print(top_products_qty)
```

```python
#Least Selling Products by Amount
least_products_amount = df.groupby("Product Name")["Total Amount"].sum().sort_values().head(10)
print("\nLeast 10 Products by Total Sales Amount:")
print(least_products_amount)
```

```python
#Visualization - Top Selling Products (Amount)
plt.figure(figsize=(10,6))
sns.barplot(x=top_products_amount.values, y=top_products_amount.index, palette="viridis")
plt.title("Top 10 Offer Products by Sales Amount")
plt.xlabel("Total Sales Amount")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
#Visualization - Top Products by Quantity
plt.figure(figsize=(10,6))
sns.barplot(x=top_products_qty.values, y=top_products_qty.index, palette="rocket")
plt.title("Top 10 Offer Products by Quantity Sold")
plt.xlabel("Quantity")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
#Average Order Value per Product
avg_order_value = df.groupby("Product Name")["Total Amount"].mean().sort_values(ascending=False).head(10)
print("Top 10 Products by Average Order Value:\n", avg_order_value)
plt.figure(figsize=(10,6))
sns.barplot(x=avg_order_value.values, y=avg_order_value.index, palette="mako")
plt.title("Top 10 Products by Average Order Value")
plt.xlabel("Average Order Value")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
# Price per Unit Analysis (Unit Price)
df["Price Per Unit"] = df["Total Amount"] / df["Qty"]
price_per_unit = df.groupby("Product Name")["Price Per Unit"].mean().sort_values(ascending=False).head(10)
print("Top 10 Products by Unit Price:\n", price_per_unit)

plt.figure(figsize=(10,6))
sns.barplot(x=price_per_unit.values, y=price_per_unit.index, palette="flare")
plt.title("Top 10 Expensive Products (Unit Price)")
plt.xlabel("Unit Price")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
# Product Contribution to Total Sales
product_contribution = df.groupby("Product Name")["Total Amount"].sum()
product_percent = (product_contribution / product_contribution.sum()) * 100
top_contributors = product_percent.sort_values(ascending=False).head(10)

print("Top 10 Products by % Contribution to Sales:\n", top_contributors)

plt.figure(figsize=(8,8))
top_contributors.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='Set3')
plt.title("Product Contribution to Total Sales (%)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
