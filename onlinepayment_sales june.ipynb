```python
# Drop Rows with Missing Key Info
df.dropna(subset=["Product Name", "Total Amount", "Qty"], inplace=True)
df
```

```python
#Total Online Payment Sales by Product
top_sales = df.groupby("Product Name")["Total Amount"].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Total Online Payment Sales:\n", top_sales)
```

```python
 #Least Performing Products by Sales
least_sales = df.groupby("Product Name")["Total Amount"].sum().sort_values().head(10)
print("Bottom 10 Products by Total Online Payment Sales:\n", least_sales)
```

```python
#Top 10 Products by Quantity Sold
top_qty = df.groupby("Product Name")["Qty"].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Quantity Sold:\n", top_qty)
```

```python
#Least Sold Products by Quantity
least_qty = df.groupby("Product Name")["Qty"].sum().sort_values().head(10)
print("Bottom 10 Products by Quantity Sold:\n", least_qty)
```

```python
#Visualize Top 10 Products by Total Amount
plt.figure(figsize=(10,6))
sns.barplot(x=top_sales.values, y=top_sales.index, palette="Blues_d")
plt.title("Top 10 Products by Online Payment Sales")
plt.xlabel("Total Amount")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
#Visualize Bottom 10 Products by Total Amount
plt.figure(figsize=(10,6))
sns.barplot(x=least_sales.values, y=least_sales.index, palette="Reds_d")
plt.title("Bottom 10 Products by Online Payment Sales")
plt.xlabel("Total Amount")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
 #Visualize Top Products by Quantity
plt.figure(figsize=(10,6))
sns.barplot(x=top_qty.values, y=top_qty.index, palette="Purples")
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Quantity")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
#Calculate Average Selling Price per Product
df["Unit Price"] = df["Total Amount"] / df["Qty"]
avg_price = df.groupby("Product Name")["Unit Price"].mean().sort_values(ascending=False).head(10)
print("Top 10 Expensive Products by Unit Price:\n", avg_price)
```

```python
#Visualize Average Price per Product
plt.figure(figsize=(12,6))
sns.barplot(x=avg_price.values, y=avg_price.index, palette="cividis")
plt.title("Top 10 Expensive Products by Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```
