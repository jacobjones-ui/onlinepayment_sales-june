```python
# Total Sales Analysis
plt.figure(figsize=(10,8))
sns.histplot(df["Total Amount"], bins=30, kde=True,color="red")
plt.title("Distribution of Total Sales Amount")
plt.xlabel("Total Sales (â‚¹)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Top-Selling Products
top_products = df.groupby("Product Name")["Total Amount"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title("Top 10 Best-Selling Products by Revenue")
plt.xlabel("Total Sales (â‚¹)")
plt.ylabel("Product Name")
plt.show()
```

```python
# Profitability Analysis
df["Profit"] = df["Total Amount"] - df["Purchase Cast"]
top_profitable_products = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_profitable_products.values, y=top_profitable_products.index, palette="magma")
plt.title("Top 10 Most Profitable Products")
plt.xlabel("Profit (â‚¹)")
plt.ylabel("Product Name")
plt.show()
```

```python
# Tax Contributions (CGST, SGST, IGST)
tax_contributions = df[["CGST Amt", "SGST Amt", "IGST Amt"]].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=tax_contributions.index, y=tax_contributions.values, palette="coolwarm")
plt.title("Total Tax Contributions")
plt.xlabel("Tax Type")
plt.ylabel("Amount (â‚¹)")
plt.show()
```

```python
# Cost vs Revenue Comparison
df["Revenue"] = df["Total Amount"]
df["Cost"] = df["Purchase Cast"]

plt.figure(figsize=(10,5))
sns.boxplot(data=df[["Revenue", "Cost"]], palette="Set2")
plt.title("Revenue vs Cost Distribution")
plt.ylabel("Amount (â‚¹)")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
# Outlier Detection using Boxplots
num_columns = ["Total Amount", "Purchase Cast", "Profit", "BV Value"]
plt.figure(figsize=(12,6))
df[num_columns].boxplot()
plt.title("Outlier Detection for Key Financial Metrics")
plt.ylabel("Value (â‚¹)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Order Value (AOV)
df["AOV"] = df["Total Amount"] / df["Qty"]
plt.figure(figsize=(8,5))
sns.histplot(df["AOV"], bins=30, kde=True, color="purple")
plt.title("Distribution of Average Order Value (AOV)")
plt.xlabel("AOV (â‚¹)")
plt.ylabel("Frequency")
plt.show()
```
