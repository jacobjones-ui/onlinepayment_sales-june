```
# Drop rows with missing key values
df.dropna(subset=["Prod.Id", "Product Name", "Total Amount"], inplace=True)
df
```

```
#total profit columns
df["Profit"] = df["Total Amount"] - df["Purchase Cast"]
df
```

```
df["Total Tax"] = df["CGST Amt"] + df["SGST Amt"] + df["IGST Amt"]
df
```

```
df.shape
```

```
 #Top 10 Products by Total Sales(not visually)
 # Grouping by product
top_10_sales = df.groupby("Product Name")["Total Amount"].sum().sort_values(ascending=False).head(10)
top_10_sales
```

```
#Visualize Top 10 Products
plt.figure(figsize=(12, 6))
top_10_sales.plot(kind='bar', color='skyblue')
plt.title("Top 10 Products by Total Sales")
plt.ylabel("Total Sales (â‚¹)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
#Profit Analysis
top_10_profit = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_10_profit.plot(kind='bar', color='seagreen')
plt.title("Top 10 Products by Profit")
plt.ylabel("Profit (â‚¹)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df[["Qty", "BV Value", "TaxableValue", "Total Tax", "Total Amount", "Profit"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Financial Metrics")
plt.tight_layout()
plt.show()
```

```
#Average profit per product
avg_profit_per_product = df.groupby("Product Name")["Profit"].mean().sort_values(ascending=False).head(10)
avg_profit_per_product
```

```
#total quantity sold per product
total_qty_per_product = df.groupby("Product Name")["Qty"].sum().sort_values(ascending=False).head(10)
total_qty_per_product
```

```
top_tax_contributors = df.groupby("Product Name")["Total Tax"].sum().sort_values(ascending=False).head(10)
top_tax_contributors
```

```
#Profit Margin (%) Calculation
df["Profit Margin %"] = ((df["Profit"]) / df["Total Amount"]) * 100
df["Profit Margin %"]
```

```
#Avg profit margin
avg_profit_margin = df.groupby("Product Name")["Profit Margin %"].mean().sort_values(ascending=False).head(10)
avg_profit_margin
```

```
# Add BV to Amount % (Business Volume to Revenue Insight)
df["BV% of Revenue"] = (df["BV Value"] / df["Total Amount"]) * 100
df["BV% of Revenue"]
```
